```python
from sklearn.datasets import load_digits, load_breast_cancer
from sklearn.cross_validation import cross_val_score
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB

import numpy as np
```

```python
digits = load_digits()
breast_cancer = load_breast_cancer()
```

```python
digits.data[:5]
```

```python
digits.target[:5]
```

```python
breast_cancer.data[:5]
```

```python
breast_cancer.target[:5]
```

```python
bern_estimator = BernoulliNB()
multinom_estimator = MultinomialNB()
gauss_estimator = GaussianNB()
```

```python
score_bern = cross_val_score(estimator=bern_estimator, X=digits.data, y=digits.target)
score_multi = cross_val_score(estimator=multinom_estimator, X=digits.data, y=digits.target)
score_gauss = cross_val_score(estimator=gauss_estimator, X=digits.data, y=digits.target)
```

```python
print(score_bern.mean())
print(score_multi.mean())
print(score_gauss.mean())

with open("answer2.txt", "w") as file1:
    file1.write(str(score_multi.mean()))
```

```python
score_bern = cross_val_score(estimator=bern_estimator, X=breast_cancer.data, y=breast_cancer.target)
score_multi = cross_val_score(estimator=multinom_estimator, X=breast_cancer.data, y=breast_cancer.target)
score_gauss = cross_val_score(estimator=gauss_estimator, X=breast_cancer.data, y=breast_cancer.target)
```

```python
print(score_bern.mean())
print(score_multi.mean())
print(score_gauss.mean())

with open("answer1.txt", "w") as file1:
    file1.write(str(score_gauss.mean()))
```
