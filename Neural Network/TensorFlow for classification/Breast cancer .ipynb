```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize
```

```python
load_data = load_breast_cancer()

data = load_data.data
target = load_data.target
```

```python
print('Shape of data:', data.shape)
```

```python
data = normalize(data)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.3)
```

# Logisitc Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
lg_model = LogisticRegression()
lg_model.fit(X_train, y_train)
```

```python
prediction = lg_model.predict(X_test)
```

```python
print(classification_report(y_test, prediction))
```

# Logistic Regression with GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
print('Logistic Regression parametrs:',LogisticRegression().get_params())
```

```python
c_list = list(np.linspace(0, 5, 20))[1:-1]
max_iter_list = [100, 500, 1000, 2000]
penalty_list = ['l2', 'l1']

parametrs = {'C':c_list, 'max_iter':max_iter_list, 'penalty':penalty_list}
```

```python
model = LogisticRegression()
```

```python
gs_clf = GridSearchCV(model, parametrs, cv=5)
```

```python
gs_clf.fit(X_train, y_train)
```

```python
print('best parametrs:', gs_clf.best_params_)
```

```python
best_model = gs_clf.best_estimator_
```

```python
prediction = best_model.predict(X_test)
```

```python
print(classification_report(y_test, prediction))
```

# BaggingClassifier

```python
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_clf = DecisionTreeClassifier(max_features=int(np.sqrt(data.shape[1])))
bagging_clf = BaggingClassifier(base_estimator=dt_clf, n_estimators=100)
```

```python
bagging_clf.fit(X_train, y_train)
```

```python
prediction = bagging_clf.predict(X_test)
```

```python
print(classification_report(y_test, prediction))
```

# RandomForestClassifier

```python
rf_cls = RandomForestClassifier(n_estimators=100)
```

```python
rf_cls.fit(X_train, y_train)
```

```python
prediction = rf_cls.predict(X_test)
```

```python
print(classification_report(y_test, prediction))
```

# TensorFlow + keras

```python
from tensorflow import keras
```

```python
inputX = X_train
inputY = y_train
```

```python
model = keras.Sequential([
    keras.layers.Flatten(),
    keras.layers.Dense(25, activation=tf.nn.relu),
    # keras.layers.Dropout(0.0),
    keras.layers.Dense(10, activation=tf.nn.relu),
    # eras.layers.Dropout(0.0),
    keras.layers.Dense(2, activation=tf.nn.sigmoid)
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model_history = model.fit(inputX, inputY, epochs=1000,
          validation_data=(X_test, y_test), verbose=2)
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)

print('Test accuracy:', test_acc)
```

```python
prediction = model.predict(X_test)

result = []
for row in prediction:
    result.append(np.argmax(row))
```

```python
print(classification_report(y_test, result))
```

```python
plt.figure(figsize=(16, 10))

plt.plot(model_history.epoch, model_history.history['val_loss'])
```

```python
from sklearn.decomposition import PCA
```

```python
PCA_transformed = PCA(n_components=2, random_state=321).fit_transform(data)
```

```python
for i, r in enumerate(PCA_transformed):
    if target[i] == 0:
        plt.scatter(r[0], r[1], c='red')
    else:
        plt.scatter(r[0], r[1], c='green')
```

```python
X_pca_train, X_pca_test, y_pca_train, y_pca_test = train_test_split(PCA_transformed, target, test_size=0.3)
```

```python
c_list = list(np.linspace(0, 5, 20))[1:-1]
max_iter_list = [100, 500, 1000, 2000]
penalty_list = ['l2', 'l1']

parametrs = {'C':c_list, 'max_iter':max_iter_list, 'penalty':penalty_list}

model = LogisticRegression()
```

```python
gs_clf = GridSearchCV(model, parametrs, cv=5)

gs_clf.fit(X_pca_train, y_pca_train)

best_model = gs_clf.best_estimator_
```

```python
for i, r in enumerate(PCA_transformed):
    if target[i] == 0:
        plt.scatter(r[0], r[1], c='red')
    else:
        plt.scatter(r[0], r[1], c='green')

plt.plot(X_pca_test, best_model.coef_ * X_pca_test, linewidth=1)
```

```python
print(classification_report(y_pca_test, best_model.predict(X_pca_test)))
```
