```python
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return np.sin(x / 5.) * np.exp(x / 10.) + 5 * np.exp(-x / 2.)
```

```python
x = np.arange(1,30,0.1)
y = f(x)
```

```python
plt.plot(x, y, '-')
plt.show()
```

```python
res1 = minimize(f, 2, method='BFGS')
res2 = minimize(f, 30, method='BFGS')
res2
```

```python
res1 = round(res1.fun, 2)
res2 = round(res2.fun, 2)
```

```python
f = open('week3task1_sub.txt', 'w')
f.write(str(res1) + ' ' + str(res2))
f.close()
```
